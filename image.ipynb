```
from __future__ import absolute_import, division, print_function, unicode_literals
from tensorflow.keras import datasets, layers, models
import numpy as np
import matplotlib.pyplot as plt

train_images, test_images, train_labels, test_labels = np.load('/content/drive/Shared drives/웹해파/data/test.npy',allow_pickle=True)#내가 만든 파일을 불러온다

train_images = train_images.reshape((429, 240 , 320, 3)) 
test_images = test_images.reshape((187, 240, 320, 3))         

train_images, test_images = train_images / 255.0, test_images / 255.0

#CNN
model = models.Sequential()
#model.add(layers.Conv2D(필터의 갯수, (필터의 크기 X, Y), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPooling2D((2, 2)))                              #2*2 최고값 풀링
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

model.add(layers.Flatten())                                         #입력값(자동 설정)을 1차원으로 변경
model.add(layers.Dense(32, activation='relu'))                      #32개의 아웃풋 활성화 함수 = relu
model.add(layers.Dense(3, activation='softmax'))                    #softmax는 최종 출력값을 결정한다.
                                                                    #softmax는 출력값의 총합이 1이다 .해당값이 나올 확률을 알수있다
model.compile(optimizer='adam',                                     #데이터와 손실함수를 바탕으로 모델의 업데이트 방법결정
              loss='sparse_categorical_crossentropy',               #손실함수 훈련하는 동안 오차를 측정
              metrics=['accuracy'])                                 #훈련 단계와 테스트 단계를 모니터링을 하기위해 사용

model.fit(train_images, train_labels, epochs=30)                    #학습 시작 epochs = 학습 횟수

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print(test_acc)

image_class = ["normal","other","semi"]
predictions = model.predict(train_images)
predictions_test = model.predict(test_images)
                                                    # 테스트 이미지에 대한 예측 결과 가져오기
plt.figure(figsize=(10, 10))
for i in range(63):                                 # 이미지 25개를 표시하기 위해 25번 반복
    plt.subplot(8,8,i+1)                            # 5 X 5 사이즈 그리드에서 이미지를 표시할 위치 지정
    plt.xticks([])                                  # X축 눈금 제거
    plt.yticks([])                                  # Y축 눈금 제거
    plt.grid(False)                                 # 눈금 제거
    plt.imshow(train_images[i], cmap=plt.cm.binary)  # 창에 띄울 이미지 추가
    plt.xlabel("{} {:2.0f}%".format(image_class[np.argmax(predictions[i])],100*np.max(predictions[i])))        # 이미지의 라벨 추가 (분류 결과)
plt.show()

for i in range(27):                                 # 이미지 25개를 표시하기 위해 25번 반복
    plt.subplot(5,6,i+1)                            # 5 X 5 사이즈 그리드에서 이미지를 표시할 위치 지정
    plt.xticks([])                                  # X축 눈금 제거
    plt.yticks([])                                  # Y축 눈금 제거
    plt.grid(False)                                 # 눈금 제거
    plt.imshow(test_images[i], cmap=plt.cm.binary)  # 창에 띄울 이미지 추가
    plt.xlabel("{} {:2.0f}%".format(image_class[np.argmax(predictions_test[i])],100*np.max(predictions_test[i])))        # 이미지의 라벨 추가 (분류 결과)
plt.show()
```
